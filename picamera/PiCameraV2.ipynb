```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import astropy
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import speclite.filters
```

## PiCamera V2: Sony IMX219

```python
v2_resp = pd.read_csv('pi-camera-response-curves/Sony_IMX219_spectral_response.csv')
```

```python
v2_resp
```

```python
plt.plot(v2_resp.wavelength, (v2_resp.blue)/100., ls='-', color='blue', label='Blue')
plt.plot(v2_resp.wavelength, (v2_resp.green)/100., ls='-', color='green', label='Green')
plt.plot(v2_resp.wavelength, (v2_resp.red)/100., ls='-', color='red', label='Red')

plt.xlabel('Wavelength (nm)')
plt.ylabel('Relative response')

plt.legend(loc='best')
plt.title('PiCamera V2 Spectral Response')
```
